```python
# CDM - JSON
import xml.etree.ElementTree as ET
import json
import os

# Define the path to the XSD file
file_path = r"C:\Users\vmd1\OneDrive - NIST\Documents\preeditbuild.xsd"

# Check if the file exists
if not os.path.exists(file_path):
    raise FileNotFoundError(f"The file {file_path} does not exist.")

try:
    # Parse the XML schema (XSD) file using ElementTree
    tree = ET.parse(file_path)
    root = tree.getroot()
    
    elements = []

    # Function to extract element attributes and nested details
    def extract_element_details(elem):
        details = {}
        # Extract common attributes
        details['Element Name'] = elem.get('name')
        details['Type'] = elem.get('type')
        details['Min Occurs'] = elem.get('minOccurs')
        details['Max Occurs'] = elem.get('maxOccurs')
        
        # Extract documentation if available
        doc_elem = elem.find(".//{http://www.w3.org/2001/XMLSchema}documentation")
        if doc_elem is not None and doc_elem.text:
            details['Documentation'] = doc_elem.text.strip()
            # Might want to change this to say N/A or something
        
        return details
    
    # Find all elements and extract their details
    for elem in root.findall(".//{http://www.w3.org/2001/XMLSchema}element"):
        element_details = extract_element_details(elem)
        elements.append(element_details)
    
    # Convert the elements list to JSON format
    json_data = json.dumps(elements, indent=2)
    
    # Print the JSON data
    print(json_data)

except ET.ParseError as e:
    print(f"XML parsing error: {e}")
except FileNotFoundError as e:
    print(e)
except PermissionError:
    print(f"Permission denied: Unable to access the file {file_path}.")
    print("Try running the script as an administrator or check file permissions.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
